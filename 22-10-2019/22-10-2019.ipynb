# matplotlib ticks
#### 1.ticks is values within coordinators

```python
import matplotlib.pyplot as p1
y=[5,10,15,20,25,30,35,40,45,50]
x=[9,8,5,6,3,2,1,4,7,10]
p1.plot(y,x)
p1.title('A graph with example  of ticks')
p1.xlabel('Natural numbers')
p1.ylabel('values with ++5 up to 50')
p1.show()
```

```python
import matplotlib.pyplot as p1
import numpy as np
p1.yticks(np.arange(10,20,1))
p1.yticks(np.arange(5,51,5))
p1.plot(x,y)
p1.title('matplotlib ticks')
p1.xlabel('10 natural numbers')
p1.ylabel('5-50 values with 5+')
p1.show()
```

### representing a sine wave
#### 1.A=sine(wt)
#### 2.A=amplitude
#### 3.w=frequency

```python
import numpy as np
import matplotlib.pyplot as p1
Time=np.arange(1,10,0.1)
Amplitude=np.sin(Time)
p1.plot(Time,Amplitude)
p1.title("Sine wave A=sin(wt)")
p1.xlabel('Time')
p1.ylabel('Amplitude')
p1.show()
```

```python
import pandas as pd 
df=pd.read_csv('income.csv')
df
```

```python
import matplotlib.pyplot as p1
years=df.columns[2::]
income=df.iloc[1,2:]
p1.plot(years,income)
p1.title("Income of 2005 t0 2013")
p1.xlabel('years')
p1.ylabel('income')
p1.savefig("2005 to 2013")
p1.show()
```

```python
#plot all states income from 2005 t0 2013
import matplotlib.pyplot as p1
years=df.columns[-5::]
allstateincome=df.iloc[0:,2:]
p1.plot(years,allstateincome)
p1.title("Income of all states")
p1.xlabel('years')
p1.ylabel('states')
p1.savefig("all states income")
p1.show()
```
